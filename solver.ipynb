A solver for the equations of motion of a double pendulum with a point mass at the end of the first rod and another point mass at the end of the second rod.

The equations of motion are given by:
$$
\begin{align*}
(m_1 + m_2) L_1 \ddot{\theta}_1 + m_2 L_2 \ddot{\theta}_2 \cos(\theta_1 - \theta_2) + m_2 L_2 \dot{\theta}_2^2 \sin(\theta_1 - \theta_2) + (m_1 + m_2) g \sin \theta_1 &= 0, \\
m_2 L_2 \ddot{\theta}_2 + m_2 L_1 \ddot{\theta}_1 \cos(\theta_1 - \theta_2) - m_2 L_1 \dot{\theta}_1^2 \sin(\theta_1 - \theta_2) + m_2 g \sin \theta_2 &= 0
\end{align*}
$$

where $\theta_1$ is the angle between the first rod and the vertical, and $\theta_2$ is the angle between the second rod and the vertical.

```python
import matplotlib.pyplot as plt
import numpy as np

import dbl_pendulum_solver

# %matplotlib widget doesn't seem to work?
%matplotlib inline
```

```python
# Constants
L1 = 0.5  # Length of pendulum 1 (meters)
L2 = 0.5  # Length of pendulum 2 (meters)
m1 = 1    # Mass of first point (kg)
m2 = 3    # Mass of second point (kg)

# Initial conditions
initial_state = dbl_pendulum_solver.State(
    theta1=np.pi / 2,  # Initial angle for pendulum 1 (radians)
    theta1_dot=0.0,    # Initial angular velocity for pendulum 1 (rad/s)
    theta2=np.pi / 4,  # Initial angle for pendulum 2 (radians)
    theta2_dot=0.0,    # Initial angular velocity for pendulum 2 (rad/s)
)

duration = 20         # Total time (seconds)
```

```python
# Solve the ODE
solution = dbl_pendulum_solver.solve(duration, initial_state, L1, L2, m1, m2)
```

```python
# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(solution.t, solution.theta1*180/np.pi, label=r'$\theta_1$ (Pendulum 1)')
plt.plot(solution.t, solution.theta2*180/np.pi, label=r'$\theta_2$ (Pendulum 2)')
plt.xlabel('Time (s)')
plt.ylabel('Angle (deg)')
plt.legend()
plt.grid(True)
plt.show()
```
